## Rolling correlation - along with multiIndexes and .groupby 

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ['date', 'industry']
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda corr: corr.values.mean())
ind_tr36corr.plot(figsize=(12, 6))
```

```python
tmi_tr36rets.plot(label='Tr36 Mo Rets', legend=True, figsize=(12, 6))
ind_tr36corr.plot(label='Tr36 Mo Corr', legend=True, secondary_y=True)
```

We see why diversification fails, just when we need it most. When the market is failling apart, that's when we really want correlation to bail us out but it goes up meaning that all assets are failling down. Now, it's not always the case but it's most of the time.

```python
tmi_tr36rets['2007':].plot(label='Tr36 Mo Rets', legend=True, figsize=(12, 6))
ind_tr36corr['2007':].plot(label='Tr36 Mo Corr', legend=True, secondary_y=True)
```
